```python
import numpy as np
import os
import pandas as pd
```

```python
os.chdir("../Prepros_training_data")
nr_deleted_frames = 10
```

## Inputs

```python
retval = os.getcwd()
print("Current working directory %s" % retval)
df_controlls = pd.read_csv('inputs.csv', header=None)

intervall = []
nr_rows = len(df_controlls)
for i in range(nr_rows-nr_deleted_frames, nr_rows):
    intervall.append(i)
df_controlls_deleted = df_controlls.drop(intervall)
df_controlls_deleted.to_csv('inputs.csv', header=False, index=False)
df_controlls, df_controlls_deleted
```

```python
retval = os.getcwd()
print("Current working directory %s" % retval)
df_matrix = pd.read_csv('screen_matrix.csv', header=None)
nr_rows = len(df_matrix)

intervall = []
nr_rows = len(df_matrix)
for i in range(nr_rows-nr_deleted_frames, nr_rows):
    intervall.append(i)
df_screen_matrix_deleted = df_matrix.drop(intervall)
df_screen_matrix_deleted.to_csv('screen_matrix.csv', header=False, index=False)
df_matrix, df_screen_matrix_deleted
```
