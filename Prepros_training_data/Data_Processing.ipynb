```python
import numpy as np
import cv2 as cv
import pandas as pd
import re
import pickle
```

# x_train

```python
df = pd.read_csv("screen_matrix.csv", header=None)
df

```

```python
def hasNumbers(inputString):
    return (re.findall(r'\d', inputString))
nr_pixels = len(df.columns)
nr_images = len(df)
all_images = []

for df_row in range(0,nr_images):
    row = []
    for pixel in range(nr_pixels):
        #print(df.loc[0,:][pixel])
        nr = hasNumbers(str(df.loc[df_row,:][pixel]))
        nr = int(nr[0])
        #print(nr)
        row.append(nr)
    all_images.append(row)
```

```python
all_images = np.array(all_images)
all_imagesRE = np.reshape(all_images, (nr_images, 16, 13))

x_train = np.zeros(shape=(nr_images, 13, 16))
for images in range(nr_images):
    b = (all_imagesRE[images].transpose())
    x_train[images] = b
```

```python
x_train[-1]
```

```python
print("Shape before saving: ", x_train.shape)
x_train_saved = np.reshape(x_train, (x_train.shape[0], x_train.shape[1]*x_train.shape[2]))
print("Shape after saving: ", x_train_saved.shape)
```

```python
with open('x_train_saved.pickle', 'wb') as f:
    pickle.dump([x_train_saved], f)
```

# y_train

```python
df_y = pd.read_csv("inputs.csv", header=None)
y_train = np.array(df_y)
y_train = df_y.drop([3, 4, 5, 6, 7], axis=1)
all_false = ((y_train[0] == 0) & (y_train[1] == 0) & (y_train[2] == 0))
all_false = pd.DataFrame(all_false, columns=['all_false'])
y_train = pd.concat([y_train, all_false], axis=1)
y_train = np.array(y_train.astype(int))

print("y_train shape: ", y_train.shape)
with open('y_train_saved.pickle', 'wb') as f:
    pickle.dump([y_train], f)
```
