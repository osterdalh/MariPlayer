```python
df_y = pd.read_csv("inputs.csv", header=None)
y_train = np.array(df_y)
y_train = df_y.drop([3, 4, 5, 6, 7], axis=1)
all_false = ((y_train[0] == 0) & (y_train[1] == 0) & (y_train[2] == 0))
all_false = pd.DataFrame(all_false, columns=['all_false'])
y_train = pd.concat([y_train, all_false], axis=1)
y_train.rename(columns={0: 'hopp', 1: 'venstre', 2: 'høyre', 'all_false': 'null'}, inplace=True)
y_train = y_train.drop(['venstre'], axis=1)
hopp_right = ((y_train['hopp'] == 1) & (y_train['høyre'] == 1))
hopp_right = pd.DataFrame(hopp_right, columns=['HH'])
y_train = pd.concat([y_train, hopp_right], axis=1)
#y_train = pd.DataFrame(y_train, columns=['hopp', 'venstre', 'hoyre', 'null'])
y_train = y_train.drop(['null'], axis=1)


y_train = np.array(y_train.astype(int))

for i in range(len(y_train)):
    if (y_train[i][2] == 1):
        y_train[i][0] = 0
        y_train[i][1] = 0
        
for i in range(len(y_train)):
    if (y_train[i][0] == 0 and y_train[i][1] == 0 and y_train[i][2] == 0):
        y_train[i][2] = 1
        
```

```python
from random import shuffle

hopp = 0
hoyre = 0
HH = 0

antall_hoyre = []

for i in range(len(y_train)):
    if y_train[i][0] == 1:
        hopp = hopp + 1
    elif y_train[i][1] == 1:
        hoyre = hoyre + 1
        antall_hoyre.append(i)
    elif y_train[i][2] == 1:
        HH = HH + 1
print("antall hopp:", hopp, "antall høyre:", hoyre, "antall HH", HH)    
    
shuffle(antall_hoyre)        
#print(antall_hoyre)
delete_frames = antall_hoyre[:hoyre-HH+500]
```

```python
df_y = pd.DataFrame(y_train)
df_y = df_y.drop(delete_frames)
y_train = np.array(df_y)

print("y_train shape: ", y_train.shape)
with open('y_train_saved.pickle', 'wb') as f:
    pickle.dump([y_train], f)
```
