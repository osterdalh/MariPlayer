```python
import numpy as np
import cv2 as cv
import pandas as pd
import re
import pickle
import keras 
from keras.models import load_model
from keras.optimizers import Adam
from keras.layers import Dense, Conv2D, Dropout, Flatten
from keras import Sequential
import matplotlib.pyplot as plt

import time
```

```python

model = load_model('mariplayer.h5')
```

```python
while True: 
    df = pd.read_csv("screen_matrix_live.csv", header=None)
    nr_images = 1
    nr_pixels = len(df.columns)
    all_images = np.array(df)
    all_imagesRE = np.reshape(all_images, (nr_images, 16, 13))

    x_train = np.zeros(shape=(nr_images, 13, 16))
    b = (all_imagesRE.transpose())
    x_train = b
    x_train = x_train.reshape(1, 13, 16,1)
    
    controll = model.predict(x_train)
    index_max = np.argmax(controll)

    file_action = open('action.txt', 'w') 
    file_action.write(str(index_max))
    file_action.close()
    print("action:", index_max, "SSH:", controll[0][index_max])


    
    file_step = open('step.txt', 'w') 
    file_step.write(str(1))
    file_step.close()
    
    time.sleep(0.1)
    
    
    
    
```
