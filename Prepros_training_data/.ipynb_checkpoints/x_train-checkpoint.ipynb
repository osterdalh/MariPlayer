```python
df = pd.read_csv("enes.csv", header=None)
df

```

```python
def hasNumbers(inputString):
    return (re.findall(r'\d', inputString))
nr_pixels = len(df.columns)
nr_images = len(df)
all_images = []

for df_row in range(200,nr_images):
    row = []
    for pixel in range(nr_pixels):
        #print(df.loc[0,:][pixel])
        nr = hasNumbers(str(df.loc[0,:][pixel]))
        #print(nr)
        row.append(nr)
    all_images.append(row)
```
